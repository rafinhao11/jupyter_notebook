```python
a = int(input("A: "))
b = int(input("B: "))
c = int(input("C: "))

delta = (b ** 2) - 4 * a * c


if a != 0:
    print("Não é equação do segundo grau!")
elif delta < 0:
    print("Não existe raiz!")
elif delta == 0:
    print("Existe raiz única")
    x = -(b ** 2) + delta * (1 / 2) / 2 * a
    print("A raiz de x é {0}".format(x))
elif delta >= 0:
    x1 = (-b ** 2 + delta * (1 / 2)) / 2 * a
    x2 = (-b ** 2 - delta * (1 / 2)) / 2 * a
    print("X' = {0} e X'' = {1}".format(x1, x2))
else:
    print("Inválido!")
```
